```python
import pandas as pd
import numpy as np

file_path=("C:\\Users\\RasNas\\Desktop\\myexcel.xlsx")

df=pd.read_excel(file_path)
df.head()

```

```python
# Replace "Height" column values with random numbers between 150 and 180
df['Height'] = np.random.randint(150, 181, df.shape[0])
df.head()
```

```python
# Distribution of employees 
team_distribution = df['Team'].value_counts()
# Calculate percentage split
team_percentage = (team_distribution / len(df)) * 100

# Combine both into a DataFrame
team_distribution_df = pd.DataFrame({'Count': team_distribution, 'Percentage': team_percentage})

team_distribution_df
# Segregation of employees 
position_distribution = df['Position'].value_counts()
position_distribution_df = pd.DataFrame({'Count': position_distribution})
position_distribution_df
```

```python
# Define age groups
bins = [20, 25, 30, 35, 40, 45, 50]
labels = ['20-24', '25-29', '30-34', '35-39', '40-44', '45-50']
# Create age group column
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
# Count the number of employees in each age group
age_group_distribution = df['Age Group'].value_counts().sort_index()
age_group_distribution_df = pd.DataFrame({'Count': age_group_distribution})

age_group_distribution_df
```

```python
# Sum salaries  team
team_salary = df.groupby('Team')['Salary'].sum().sort_values(ascending=False)

# Sum salaries  position
position_salary = df.groupby('Position')['Salary'].sum().sort_values(ascending=False)

team_salary_df = pd.DataFrame({'Total Salary': team_salary})
position_salary_df = pd.DataFrame({'Total Salary': position_salary})

team_salary_df, position_salary_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot of Age vs Salary
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='Salary')
plt.title('Correlation between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
           
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x=team_distribution_df.index, y=team_distribution_df['Count'])
plt.xticks(rotation=90)
plt.title('Distribution of Employees Across Teams')
plt.xlabel('Teams')
plt.ylabel('Number of Employees')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df['Position'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Distribution of Employees Based on Positions')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=age_group_distribution_df.index, y=age_group_distribution_df['Count'])
plt.title('Distribution of Employees Across Age Groups')
plt.xlabel('Age Groups')
plt.ylabel('Number of Employees')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x=team_salary_df.index, y=team_salary_df['Total Salary'])
plt.xticks(rotation=90)
plt.title('Total Salary Expenditure by Team')
plt.xlabel('Teams')
plt.ylabel('Total Salary')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=position_salary_df.index, y=position_salary_df['Total Salary'])
plt.title('Total Salary Expenditure by Position')
plt.xlabel('Positions')
plt.ylabel('Total Salary')
plt.show()
```
